```python
import numpy as np
import pandas as pd
# 结合 numpy 创建一个 (3, 10) 区间内的包含 5 个随机整数的一维数组 ndarray 为 data1，并以此 data 再创建索引为 a, b, c, d, e 的 Series 为 s1
data1 = np.random.randint(3,10,5)
s1 = pd.Series(data1,index=['a','b','c','d','e'])
data1,s1
```

```python
# 结合 numpy 创建一个 (5, 15) 区间内，形状为 5 行 3 列的随机整数数组为 data2，并以此为基础创建 index 为 ['a', 'c', 'e', 'f', 'h']，columns 为 ['one', 'two', 'three'] 的 DataFrame，并命名为 df_test（阅读参考文章④）
data2 = np.random.randint(5,15,(5,3))
df_test = pd.DataFrame(data2,index=['a', 'c', 'e', 'f', 'h'],columns=  ['one', 'two', 'three'])
df_test
```

```python
# 运用 loc 索引以及替换的知识，按下列要求对 df_test 进行增、改、替操作。完成上述要求后，将该数据命名为 df_change
from numpy import NaN
df_test.loc["a","one"] = NaN
df_test.loc["c","two"] = -99
df_test.loc["c","three"] = -99
df_test.loc["a","two"] = -100
df_test.loc[:,"four"] = "test"
df_change = df_test.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
df_change
```

```python
# 删除 df_change 中存在缺失值的所有行
df_change.dropna(how="any")
```

```python
# 删除 df_change 中所有值都为 NaN 值的行
df_change.dropna(how="all")
```

```python
# df_change 中的所有缺失值（即 NaN）以 0 填充
df_change = df_change.fillna(0)
df_change
```

```python
# 删除 df_change中的重复行
df_change = df_change.drop_duplicates()
df_change
```
